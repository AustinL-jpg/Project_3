```python
data_mock = pd.DataFrame(data_2)

stock = []
count = 0
reset = random.randint(5, 15)

for i in data_mock['make']:
    if count == reset:
        stock.append(random.randint(1, 9))
        count = 0
        reset = random.randint(5, 15)
    else:
        stock.append(random.randint(0, 3))
    count += 1
data_mock['Stock'] = stock
data_mock.head()

vin = []
makes = []
models = []


for i, j, s, h in zip(data_mock['make'], data_mock['model'], data_mock['Stock'], data_mock['id']):
    make = i[0]
    model = j[0]
    make1 = i
    model1 = j
    num = s
    id = h
    for row in range(num):
        number = np.random.randint(100001, 999999)
        cond = np.random.choice(['Used', 'New'])
        vin_num = f'{make}{model}{number}'
        
        if cond == 'Used':
            mileage = np.random.randint(501, 150000)
        elif cond == 'New':
            mileage = np.random.randint(0, 500)

        
        vin.append([vin_num, make1, model1, cond, mileage, id])

        
    

vin = pd.DataFrame(vin, columns=['VIN', 'make', 'model', 'Condition', 'Mileage', 'id'])
vin.head()
```

```python
vin.info()


```

```python
data_clean = pd.merge(vin, data_mock, how='right', on= ('id', 'make', 'model'))
data_clean.dropna(subset=['VIN'], inplace=True)
data_clean.drop_duplicates(subset='VIN', keep='first', inplace=True)
data_clean = data_clean.rename(columns={'Make_x': 'Make'})
data_clean.to_csv('out.csv', index=False)
```
